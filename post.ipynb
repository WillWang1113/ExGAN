```python
static_sample = torch.load("ExGAN0.05_pv.pt").detach().cpu().numpy().squeeze()[:100]
# print(static_sample[0])
print((np.trapz(static_sample, dx=1/4)/10).max())

fig, axs = plt.subplots(10, 10, sharex=True, sharey=True, figsize=[10,10])
axs = axs.flatten()
for i in range(len(static_sample)):
    axs[i].plot(range(96), static_sample[i])
    axs[i].xaxis.set_major_locator(plt.MaxNLocator(2))
    axs[i].yaxis.set_major_locator(plt.MaxNLocator(2))
```
