```python
import torch
import matplotlib.pyplot as plt
```

```python
static_sample = torch.load("real_wind.pt").detach().cpu().numpy().squeeze()[:100]
# print(static_sample[0])
fig, axs = plt.subplots(10, 10, sharex=True, sharey=True)
axs = axs.flatten()
for i in range(len(static_sample)):
    axs[i].plot(range(96), static_sample[i])
    axs[i].xaxis.set_major_locator(plt.MaxNLocator(2))
    axs[i].yaxis.set_major_locator(plt.MaxNLocator(2))
```
